## Changing style and palette

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Sample survey data
data = {
    'Parents Advice': [
        "Sometimes", "Often", "Rarely", "Sometimes", "Sometimes", "Rarely", "Often", "Always",
        "Always", "Sometimes", "Rarely", "Never", "Often", "Sometimes",
        "Often", "Never", "Sometimes", "Rarely", "Often"
    ]
}

survey_data = pd.DataFrame(data)

# Change the color palette to "RdBu"
sns.set_style("whitegrid")
sns.set_palette("RdBu")

# Set order of categories
category_order = ["Never", "Rarely", "Sometimes", "Often", "Always"]

# Create a count plot of survey responses
sns.catplot(
    x="Parents Advice", 
    data=survey_data, 
    kind="count", 
    order=category_order
)

# Show plot
plt.show()
```

## Changing context scale

```python
# Change the context to "poster"
sns.set_context("poster")

# Create bar plot
sns.catplot(
    y="Parents Advice", 
    data=survey_data, 
    kind="count", 
    order=category_order
)

# Show plot
plt.show()
```

## Custom color palette

```python
# Set the style to "darkgrid"
sns.set_style("darkgrid")

# Set a custom color palette
color_palette = ["#39A7D0", "#36ADA4"]
sns.set_palette(color_palette)

# Create the box plot of age distribution by gender
sns.catplot(
    y="Parents Advice", 
    data=survey_data, 
    kind="count", 
    order=category_order
)

# Show plot
plt.show()
```

## Axessubplot vs Facetgrid
- Facetgrid is for multiple subplots such as catplot(), relplot() whereas axessubplot is for single subplot such as countplot(), scatterplot().

```python
g = sns.catplot(
    y="Parents Advice", 
    data=survey_data, 
    kind="count", 
    order=category_order
)

print(type(g))
```

```python
# Adding a title to facetgrid object

g = sns.catplot(
    y="Parents Advice", 
    data=survey_data, 
    kind="count", 
    order=category_order
)

g.fig.suptitle("Parent's advice")
```

## Customize axessubplot

```python
sns.set_context("notebook")
data = {
    "model_year": [70, 70, 71, 71, 72, 72, 73, 73, 74, 74],
    "origin": ["USA", "Europe", "USA", "Europe", "USA", "Europe", "USA", "Europe", "USA", "Europe"],
    "mpg_mean": [17.0, 22.5, 18.5, 24.0, 19.2, 25.3, 18.8, 26.1, 17.5, 25.8]
}

mpg_mean = pd.DataFrame(data)

# Create line plot
g = sns.lineplot(x="model_year", y="mpg_mean", 
                 data=mpg_mean,
                 hue="origin")

# Add a title "Average MPG Over Time"
g.set_title("Average MPG Over Time")

# Add x-axis and y-axis labels
g.set(xlabel="Car Model Year",
      ylabel="Average MPG")

# Show plot
plt.show()
```

```python
mpg = sns.load_dataset("mpg")

# Create point plot
sns.catplot(x="origin", 
            y="acceleration", 
            data=mpg, 
            kind="point", 
            join=False, 
            capsize=0.1)

# Rotate x-tick labels
plt.xticks(rotation=90)

# Show plot
plt.show()
```

```python
# Sample survey_data
survey_data = pd.DataFrame({
    "Gender": ["Male", "Female", "Female", "Male", "Female", "Male", "Female", "Male", "Female", "Male"],
    "Age": [18, 20, 19, 21, 22, 23, 20, 19, 21, 22],
    "Interested in Pets": ["Yes", "No", "Yes", "Yes", "No", "Yes", "No", "Yes", "No", "Yes"]
})

# Set palette to "Blues"
sns.set_palette("Blues")

# Adjust to add subgroups based on "Interested in Pets"
g = sns.catplot(x="Gender",
                y="Age", data=survey_data, 
                kind="box", hue="Interested in Pets")

# Set title to "Age of Those Interested in Pets vs. Not"
g.fig.suptitle("Age of Those Interested in Pets vs. Not", y=1.03)

# Show plot
plt.show()
```

```python
# Sample survey_data
survey_data = pd.DataFrame({
    "Village - town": ["Village", "Town", "Town", "Village", "Town", "Village", "Town", "Village", "Town", "Village"],
    "Likes Techno": [45, 60, 55, 40, 65, 42, 70, 38, 62, 41],
    "Gender": ["Male", "Male", "Female", "Female", "Male", "Female", "Female", "Male", "Female", "Male"]
})

# Set the figure style to "dark"
sns.set_style("dark")

# Create the bar plot with subplots per gender
g = sns.catplot(x="Village - town", y="Likes Techno", 
                data=survey_data, kind="bar",
                col="Gender")

# Add title and axis labels
g.fig.suptitle("Percentage of Young People Who Like Techno", y=1.02)
g.set(xlabel="Location of Residence", 
      ylabel="% Who Like Techno")

# Show plot
plt.show()
```
