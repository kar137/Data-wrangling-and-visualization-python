## Creating subplots with col and row

```python
# Import required libraries
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Sample dataset similar to student_data
student_data = pd.DataFrame({
    "absences": [0, 3, 4, 2, 1, 6, 8, 5, 7, 9],
    "G3": [15, 14, 13, 16, 15, 10, 9, 11, 8, 7],
    "study_time": ["<2 hours", "2 to 5 hours", "<2 hours", "2 to 5 hours", 
                   ">5 hours", "2 to 5 hours", ">5 hours", "<2 hours", 
                   "2 to 5 hours", ">5 hours"]
})

# Create scatter plots arranged in rows by study_time
sns.relplot(x="absences", y="G3", 
            data=student_data,
            kind="scatter", 
            row="study_time")

# Show plot
plt.show()
```

## Creating two-factor subplots

```python
# Sample dataset
student_data = pd.DataFrame({
    "G1": [10, 12, 13, 14, 9, 8, 15, 11, 13, 10],
    "G3": [11, 13, 14, 15, 9, 7, 16, 12, 14, 11],
    "schoolsup": ["yes", "no", "yes", "no", "yes", "no", "yes", "no", "no", "yes"],
    "famsup": ["yes", "no", "yes", "no", "yes", "no", "no", "yes", "no", "yes"]
})

# Facet scatter plot
sns.relplot(x="G1", y="G3", 
            data=student_data,
            kind="scatter", 
            col="schoolsup",
            col_order=["yes", "no"],
            row="famsup",
            row_order=["yes", "no"])

# Show plot
plt.show()
```

## Changing the size of scatter plot points

```python
import seaborn as sns
mpg = sns.load_dataset("mpg")

# Import Matplotlib and Seaborn
import matplotlib.pyplot as plt
import seaborn as sns

# Create scatter plot of horsepower vs. mpg
sns.relplot(x="horsepower", y="mpg", 
            data=mpg, kind="scatter", 
            size="cylinders",
            hue="cylinders")

# Show plot
plt.show()
```

## Changing the style of scatter plot points

```python
# Create a scatter plot of acceleration vs. mpg
sns.relplot(x="acceleration", y="mpg", kind="scatter", data=mpg, hue="origin", style="origin")



# Show plot
plt.show()
```
