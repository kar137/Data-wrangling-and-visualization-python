## Visualizing with Pandas and Seaborn

```python
# Import Matplotlib, pandas, and Seaborn
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

# Create a sample DataFrame
data = {
    "Spiders": ["Yes", "No", "Yes", "No", "Yes", "Yes", "No", "No", "Yes", "No", 
                "Yes", "No", "No", "Yes", "No", "Yes", "Yes", "No", "No", "Yes"]
}
df = pd.DataFrame(data)

# Create a count plot with "Spiders" on the x-axis
sns.countplot(x="Spiders", data=df)

# Display the plot
plt.show()
```

## Categorizing Data with Hue in Scatter Plots
Seaborn’s hue parameter allows us to introduce a third dimension to scatter plots by visually distinguishing different groups within the data using color. This is especially helpful when we want to compare relationships between two variables across categories—such as gender, region, or any class-based label.

Using hue, we can overlay multiple subgroups on the same plot without cluttering the visualization, enabling more insightful analysis from a single figure.

```python
# Create a new dataset
exam_data = pd.DataFrame({
    "study_hours": [1, 2, 3, 4, 5, 6, 2.5, 3.5, 4.5, 5.5],
    "final_score": [60, 65, 70, 75, 80, 85, 68, 73, 78, 83],
    "school_type": ["Public", "Public", "Public", "Public", "Public",
                    "Private", "Private", "Private", "Private", "Private"]
})

# Create scatter plot with school_type as hue and specific legend order
sns.scatterplot(x="study_hours", y="final_score",
                data=exam_data,
                hue="school_type",
                hue_order=["Private", "Public"])

# Show plot
plt.show()
```

```python
# Create a new dataset
exam_data = pd.DataFrame({
    "study_hours": [1, 2, 3, 4, 5, 6, 2.5, 3.5, 4.5, 5.5],
    "final_score": [60, 65, 70, 75, 80, 85, 68, 73, 78, 83],
    "school": ["School A", "School A", "School A", "School A", "School A",
               "School B", "School B", "School B", "School B", "School B"],
    "location": ["Urban", "Urban", "Urban", "Urban", "Urban",
                 "Rural", "Rural", "Rural", "Rural", "Rural"]
})

# Create a dictionary mapping subgroup values to colors
palette_colors = {"Rural": "green", "Urban": "blue"}

# Create a count plot of school with location subgroups
sns.countplot(x="school", data=exam_data, hue="location", palette=palette_colors)

# Display plot
plt.show()
```
