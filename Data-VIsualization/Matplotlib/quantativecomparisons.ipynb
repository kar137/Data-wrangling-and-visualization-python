## ðŸ“Š Bar Chart Visualization
A bar chart is used to represent categorical data using rectangular bars. The length or height of each bar reflects the magnitude of the value associated with that category. This type of chart is ideal for comparing quantities across discrete groups.

In this example, you will create a bar chart to display the number of products sold by different store locations, using data from a sales_data DataFrame. The DataFrame has store names as the index and a column named "Total_Sales" that holds the number of products sold at each location.

```python
import matplotlib.pyplot as plt
import pandas as pd

medals = pd.read_csv(r'C:\Users\Dell\OneDrive\Desktop\KaranCodes\Goal-to-be-ML-Engineer-2026\Data-wrangling-and-visualization-python\Data-VIsualization\Matplotlib\dataset\olympics_medals_2018.csv', index_col=0)

fig, ax = plt.subplots()

# Plot a bar-chart of gold medals as a function of country
ax.bar(medals.index, medals['Gold'])

# Set the x-axis tick labels to the country names
ax.set_xticklabels(medals.index, rotation=90)

# Set the y-axis label
ax.set_ylabel('Number of medals')

plt.show()
```

## ðŸ“Š Stacked Bar Chart
A stacked bar chart is useful when you want to visualize the composition of multiple sub-categories within a main category. Instead of placing bars side by side, each component is stacked vertically (or horizontally), allowing you to see both the total and the individual contributions of each variable.

```python
fig, ax = plt.subplots()

# Add bars for "Gold" with the label "Gold"
ax.bar(medals.index, medals['Gold'], label="Gold")

# Stack bars for "Silver" on top with label "Silver"
ax.bar(medals.index, medals['Silver'], bottom=medals['Gold'], label="Silver")

# Stack bars for "Bronze" on top of that with label "Bronze"
ax.bar(medals.index, medals['Bronze'], bottom=medals['Gold'] + medals['Silver'], label="Bronze")

ax.set_xticklabels(medals.index, rotation=90)
# Display the legend
ax.legend()

plt.show()
```

## ðŸ“Š Creating Histograms
A histogram is a powerful tool for visualizing the distribution of a numeric variable. It groups values into bins and shows how many data points fall into each bin, giving insights into the shape, spread, and center of the data.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Generate synthetic data
np.random.seed(42)

# Computer Science students' scores
cs_scores = np.random.normal(loc=78, scale=10, size=100)
cs_students = pd.DataFrame({'Final_Score': np.clip(cs_scores, 0, 100)})

# Mechanical Engineering students' scores
me_scores = np.random.normal(loc=72, scale=12, size=100)
me_students = pd.DataFrame({'Final_Score': np.clip(me_scores, 0, 100)})

# Plotting
fig, ax = plt.subplots()

# Plot histogram of final exam scores for Computer Science students
ax.hist(cs_students['Final_Score'], alpha=0.7, label='Computer Science')

# Compare with histogram of final exam scores for Mechanical Engineering students
ax.hist(me_students['Final_Score'], alpha=0.7, label='Mechanical Engineering')

# Set x-axis label
ax.set_xlabel("Final Exam Score")

# Set y-axis label
ax.set_ylabel("Number of Students")

# Add a legend
ax.legend()

plt.show()
```

## ðŸ“Š Adding Error Bars to a Bar Chart
Error bars provide insight into the variability or uncertainty of data. They help to assess whether the differences between groups are meaningful or could be due to natural variation.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Create synthetic data
np.random.seed(42)

# Simulated heights for men's rowing (typically taller)
mens_rowing = pd.DataFrame({
    'Height': np.random.normal(loc=190, scale=5, size=100)  # Mean ~190 cm
})

# Simulated heights for men's gymnastics (typically shorter)
mens_gymnastics = pd.DataFrame({
    'Height': np.random.normal(loc=168, scale=6, size=100)  # Mean ~168 cm
})

# Plotting
fig, ax = plt.subplots()

# Bar for rowing: mean height with std deviation as error
ax.bar("Rowing", mens_rowing['Height'].mean(), yerr=mens_rowing['Height'].std())

# Bar for gymnastics: mean height with std deviation as error
ax.bar("Gymnastics", mens_gymnastics['Height'].mean(), yerr=mens_gymnastics['Height'].std())

# Label the y-axis
ax.set_ylabel("Height (cm)")

plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Define months
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun",
          "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]

# Create synthetic Seattle weather data
seattle_weather = pd.DataFrame({
    "MONTH": months,
    "MLY-TAVG-NORMAL": [40, 43, 47, 52, 58, 63, 67, 67, 62, 53, 46, 41],
    "MLY-TAVG-STDDEV": [4, 3, 3, 2, 2, 2, 2, 2, 2, 3, 3, 4]
})

# Create synthetic Austin weather data
austin_weather = pd.DataFrame({
    "MONTH": months,
    "MLY-TAVG-NORMAL": [50, 54, 61, 69, 76, 82, 85, 86, 80, 70, 60, 52],
    "MLY-TAVG-STDDEV": [3, 3, 3, 2, 2, 2, 2, 2, 2, 2, 3, 3]
})

# Plotting
fig, ax = plt.subplots()

# Add Seattle temperature data in each month with error bars
ax.errorbar(seattle_weather['MONTH'],
            seattle_weather['MLY-TAVG-NORMAL'],
            seattle_weather['MLY-TAVG-STDDEV'],
            label='Seattle')

# Add Austin temperature data in each month with error bars
ax.errorbar(austin_weather['MONTH'],
            austin_weather['MLY-TAVG-NORMAL'],
            austin_weather['MLY-TAVG-STDDEV'],
            label='Austin')

# Set the y-axis label
ax.set_ylabel("Temperature (Fahrenheit)")

# Add a legend
ax.legend()

plt.show()
```

## ðŸ“¦ Visualizing Score Distributions with Boxplots
Boxplots are a great way to summarize the distribution of a dataset. They show the median, interquartile range (IQR), and the range where most of the data lies. Any outliers that fall far outside the typical spread are shown as individual points.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Create synthetic data
np.random.seed(42)

# Simulated heights for men's rowing (typically taller)
mens_rowing = pd.DataFrame({
    'Height': np.random.normal(loc=190, scale=5, size=100)  # mean 190 cm
})

# Simulated heights for men's gymnastics (typically shorter)
mens_gymnastics = pd.DataFrame({
    'Height': np.random.normal(loc=168, scale=6, size=100)  # mean 168 cm
})

# Plotting
fig, ax = plt.subplots()

# Boxplot of heights for both groups
ax.boxplot([mens_rowing['Height'], mens_gymnastics['Height']])

# Set x-axis labels
ax.set_xticklabels(["Rowing", "Gymnastics"])

# Set y-axis label
ax.set_ylabel("Height (cm)")

plt.show()
```
