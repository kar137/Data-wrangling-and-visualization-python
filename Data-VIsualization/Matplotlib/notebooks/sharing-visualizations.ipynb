## Styling Plots

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("../Matplotlib/dataset/stockdata.csv")

plt.style.use("ggplot")
fig, ax = plt.subplots()

# Add data: "co2", "relative_temp" as x-y, index as color
ax.scatter(df["close_price"], df["volume"], c=df.index)

# Set the x-axis label to "CO2 (ppm)"
ax.set_xlabel("Closing Price")

# Set the y-axis label to "Relative temperature (C)"
ax.set_ylabel("Volume")

plt.show()
```

## Saving a Visualization Multiple Times
To share your visual insights or include them in reports or presentations, it's essential to save the plot as a file. Matplotlib allows you to do this using the savefig() method of the Figure object. You can save the same plot multiple times in different formats (e.g., PNG, PDF, SVG) depending on your need.

```python
import numpy as np

# Create synthetic dataset
np.random.seed(1)
study_hours = np.random.uniform(1, 10, 100)
exam_scores = 5 * study_hours + np.random.normal(0, 5, 100)

students = pd.DataFrame({
    "study_hours": study_hours,
    "exam_score": exam_scores
})

# Plotting
fig, ax = plt.subplots()

# Add data: "study_hours" on x-axis, "exam_score" on y-axis
ax.scatter(students["study_hours"], students["exam_score"])

# Set the x-axis label to "Study Hours"
ax.set_xlabel("Study Hours")

# Set the y-axis label to "Final Exam Score"
ax.set_ylabel("Final Exam Score")
```

## Exporting Figures in Custom Dimensions
When saving a figure, it's often useful to control its physical dimensions—especially if it will be used in reports, presentations, or publications. Matplotlib allows you to set the size of a figure using the set_size_inches() method from the Figure object. This method takes two arguments: the width and height of the figure in inches, helping ensure your output is properly scaled for its final use.

```python
fig, ax = plt.subplots()

# Add data: "co2", "relative_temp" as x-y, index as color
ax.scatter(df["close_price"], df["volume"], c=df.index)

# Set the x-axis label to "CO2 (ppm)"
ax.set_xlabel("Closing Price")

# Set the y-axis label to "Relative temperature (C)"
ax.set_ylabel("Volume")

fig.set_size_inches([5, 3])
fig.savefig("stock.jpg")
```

## Automatically Adapting to Column Categories
A powerful feature of Matplotlib is its ability to dynamically adjust to different datasets. For instance, when dealing with categorical data, you might not know ahead of time how many unique categories exist. Using Python and Matplotlib, you can automatically retrieve these unique values and generate visualizations (like bar charts) that adapt to them—ensuring flexibility and scalability when visualizing real-world datasets.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample dataset: Student study hours by major
data = {
    "Major": ["CS", "ME", "CE", "CS", "CE", "ME", "EE", "CS", "EE", "ME", "CE", "CS", "EE", "ME", "CE", "EE"],
    "Study_Hours": [35, 30, 28, 36, 27, 29, 31, 38, 30, 33, 25, 37, 32, 34, 26, 33]
}
df = pd.DataFrame(data)

# Unique majors
majors = df["Major"].unique()

# Plotting
fig, ax = plt.subplots()

# Loop through majors
for major in majors:
    major_df = df[df["Major"] == major]
    ax.bar(major, major_df["Study_Hours"].mean(), yerr=major_df["Study_Hours"].std())

ax.set_ylabel("Average Study Hours")
ax.set_xticklabels(majors, rotation=90)

# Save figure
fig.savefig("study_hours_by_major.png")
```
